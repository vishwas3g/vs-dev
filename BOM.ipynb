```python
# prompt: code to search through merged_df by multipal  component as user input  and get relevent data related

def search_merged_df(merged_df_subset, components):
  """
  Searches the merged_df for rows that contain all of the specified components.

  Args:
    merged_df: The merged dataframe.
    components: A list of components to search for.

  Returns:
    A dataframe containing the rows that contain all of the specified components.
  """

  # Convert the components list to a set for faster lookup.
  components_set = set(components)

  # Filter the dataframe to only include rows that contain all of the components.
  filtered_df = merged_df_subset[merged_df_subset.apply(lambda row: components_set.issubset(set(row)), axis=1)]

  return filtered_df

# Get user input for components to search for.
components = input("Enter components to search for, separated by commas: ").split(",")

# Search the merged_df for rows that contain all of the specified components.
# Call the correctly named function 'search_merged_df'
filtered_df = search_merged_df(merged_df_subset, components)

# Display the filtered dataframe.
print(filtered_df)

# Download the filtered dataframe as a CSV file.
filtered_df.to_csv('filtered_df.csv')
files.download('filtered_df.csv')
```
